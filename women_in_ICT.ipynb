```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'The participation rate of women in ICT.csv'
df = pd.read_csv(file_path)

# Clean and prepare the data
df_clean = df[['السنة', 'القيمة']].dropna()
df_clean['القيمة'] = df_clean['القيمة'].str.replace('%', '').astype(float)  # float for percentages
df_clean['السنة'] = df_clean['السنة'].astype(int)  # int for years
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df_clean['السنة'], df_clean['القيمة'], marker='o')

# Force x-axis to show only the years as integers
plt.xticks(df_clean['السنة'])

plt.title('Participation Rate of Women in ICT (Saudi Arabia)')
plt.xlabel('Year')
plt.ylabel('Participation Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'The participation rate of women in ICT.csv'
df = pd.read_csv(file_path)

# Clean and prepare the data
df_clean = df[['السنة', 'القيمة']].dropna()
df_clean['القيمة'] = df_clean['القيمة'].str.replace('%', '').astype(float)
df_clean['السنة'] = df_clean['السنة'].astype(int)

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(df_clean['السنة'], df_clean['القيمة'], width=0.6)
plt.title('Participation Rate of Women in ICT (Saudi Arabia)')
plt.xlabel('Year')
plt.ylabel('Participation Rate (%)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```
