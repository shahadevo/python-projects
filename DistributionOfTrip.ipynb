```python
import pandas as pd
# Load the dataset
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('nyc_taxis.csv')
print(df.head())
print(df.columns)

filtered_df = df[df['trip_distance'] < 50]


# Plotting the histogram of trip distances
plt.hist(df['trip_distance'], bins=20, edgecolor='black', linewidth=0.25)
x_lab="trip distance (miles)"
y_lab="number of trips"
plt.xlabel(x_lab)
plt.ylabel(y_lab)   
plt.title("Distribution of trip distance in NYC taxis")
plt.grid (True)

plt.show()
```
